```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cleaned_data.csv')
```

```python
print(data)
```

#FrailtyPercentageInTheGivenDataSet

```python
frailty_counts = data['Frailty'].value_counts()
plt.pie(frailty_counts, labels=['No Frailty', 'Frailty'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

#Frailty Percentage by Age

```python
# Group the data by age and calculate the percentage of frailty status 1 at each age
age_frailty_percentage = data.groupby('Age (years)')['Frailty'].mean() * 100

plt.figure(figsize=(6, 3))
age_frailty_percentage.plot(kind='bar')
plt.title('Frailty Percentage by Age')
plt.xlabel('Age (years)')
plt.ylabel('Frailty Percentage (%)')
plt.show()
```

#Grip Strength Distribution by Age

```python
plt.figure(figsize=(6, 3))
plt.scatter(data['Age (years)'], data['Grip Strength (kg)'])
plt.title('Grip Strength Distribution by Age')
plt.xlabel('Age (years)')
plt.ylabel('Grip Strength (kg)')
plt.show()
```
